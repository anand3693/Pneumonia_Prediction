```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
```

```python
model = load_model("pneumonia_model.h5")
```

```python
img_path = r"C:\Users\lenovo\Desktop\Phnumonia Model\download.jpeg"  # path to your new image

# Load image
img = image.load_img(r"C:\Users\lenovo\Desktop\Phnumonia Model\download.jpeg", target_size=(150, 150))  # resize

# Convert to array
img_array = image.img_to_array(img)

# Expand dims to make it (1, 150, 150, 3) for model
img_array = np.expand_dims(img_array, axis=0)

# Normalize like training
img_array /= 255.0
```

```python
prediction = model.predict(img_array)

if prediction[0][0] > 0.5:
    print("Predicted: Pneumonia")
else:
    print("Predicted: Normal")
```
